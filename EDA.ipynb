# MONTE CARLO

Simulación Monte Carlo

La simulación Monte Carlo utiliza muestreo aleatorio
para modelar incertidumbre en sistemas complejos.

El método permite ejecutar miles de escenarios
posibles bajo condiciones variables.

### Aplicaciones en Agricultura Inteligente

- predicción de consumo,
- riesgo de sequía,
- estrés hídrico,
- ahorro esperado,
- eficiencia operativa.

### Beneficios Analíticos

Monte Carlo permite estimar:

- probabilidades,
- intervalos de confianza,
- riesgos operativos,
- comportamiento futuro del sistema.

### Interpretación

La simulación probabilística ayuda a construir
estrategias preventivas para optimizar
recursos hídricos.

```python

n_sim = 10000

media = metricas[
    'ahorro_litros'
].mean()

std = metricas[
    'ahorro_litros'
].std()

resultados = []

for i in range(n_sim):

    valor = np.random.normal(
        media,
        std
    )

    resultados.append(valor)

resultados = np.array(
    resultados
)
```

```python
# HISTOGRAMA MONTE CARLO

sns.histplot(
    resultados,
    kde=True
)

plt.title("Monte Carlo")

plt.show()
```

La gráfica Monte Carlo muestra los resultados de miles de simulaciones aleatorias. La mayoría de los valores se concentra cerca de 0 y 0.5e7, formando una curva parecida a una campana. Esto indica que los resultados promedio son los más probables, mientras que los valores muy bajos o muy altos ocurren pocas veces.

```python
# INTERVALOS

ic = np.percentile(
    resultados,
    [2.5,97.5]
)

print(ic)
```

```python
# VAR
VaR_95 = np.percentile(
    resultados,
    5
)

print(VaR_95)
```

# CADENAS DE MARKOV
Cadenas de Markov

Las cadenas de Markov modelan sistemas
dinámicos mediante probabilidades de transición.

El principio Markoviano establece que
el siguiente estado depende únicamente
del estado actual.

### Aplicación en Riego Inteligente

Permite modelar estados como:

- suelo seco,
- humedad óptima,
- saturación,
- fallas operativas.

### Matriz de Transición

Representa probabilidades de evolución
del sistema entre estados operativos.

### Objetivo

Evaluar estabilidad, persistencia
y riesgo operacional.

```python

def estado(h):

    if h < -1:
        return 'SECO'

    elif h < 0:
        return 'OPTIMO'

    elif h < 1:
        return 'HUMEDO'

    else:
        return 'SATURADO'
```

```python
# ESTADOS

sensores['estado'] = sensores[
    'humedad'
].apply(estado)


# MATRIZ TRANSICIÓN

transiciones = pd.crosstab(
    sensores['estado'][:-1],
    sensores['estado'][1:],
    normalize='index'
)

display(transiciones)

# HEATMAP MARKOV

sns.heatmap(
    transiciones,
    annot=True
)

plt.show()
```

El heatmap muestra que cada estado de humedad (SECO, OPTIMO, HUMEDO y SATURADO) permanece igual con probabilidad 1, y no existen transiciones entre estados diferentes. Esto indica que el sistema se mantuvo completamente estable durante la simulación.


# DETECCIÓN DE ANOMALÍAS
Detección de Anomalías

Las anomalías representan comportamientos
estadísticamente inusuales.

En IoT agrícola pueden indicar:

- fugas,
- sensores dañados,
- errores de transmisión,
- sobreconsumo,

### Isolation Forest

Isolation Forest utiliza aislamiento recursivo
para detectar observaciones anómalas
en espacios multidimensionales.

### Objetivo

Detectar riesgos operativos
y prevenir desperdicio hídrico.

```python

z = np.abs(
    stats.zscore(
        sensores[['humedad','temperatura']]
    )
)

anomalias = sensores[
    (z > 3).any(axis=1)
]

display(anomalias.head())

# ISOLATION FOREST

modelo_if = IsolationForest(
    contamination=0.02
)

sensores['anomalia'] = modelo_if.fit_predict(
    sensores[['humedad','temperatura']]
)

# VISUALIZACIÓN ANOMALÍAS

sns.scatterplot(
    data=sensores,
    x='temperatura',
    y='humedad',
    hue='anomalia'
)

plt.show()
```

La gráfica muestra anomalías detectadas entre temperatura y humedad. Los puntos morados (1) representan datos normales y los puntos claros (-1) representan anomalías o valores raros. La mayoría de los datos son normales, mientras que las anomalías aparecen principalmente en valores extremos de humedad cercanos a -1.9 y temperaturas alrededor de ±1.7. Esto indica que el sistema detectó pocos comportamientos fuera de lo común.


# SERIES TEMPORALES
Forecasting Temporal

El forecasting temporal permite anticipar
comportamientos futuros del sistema hídrico.

### Modelo ARIMA

ARIMA modela:

- autocorrelación,
- tendencia,
- diferenciación temporal.

### Aplicaciones

- predicción de consumo,
- comportamiento hídrico,

```python

resultado_adf = adfuller(
    diario.dropna()
)

print(resultado_adf)

# ARIMA

modelo = ARIMA(
    diario.dropna(),
    order=(1,1,1)
)

resultado_arima = modelo.fit()

print(
    resultado_arima.summary()
)

# FORECAST

forecast = resultado_arima.forecast(
    steps=30
)

plt.plot(diario)

plt.plot(forecast)

plt.show()
```

La gráfica muestra una comparación de valores en el tiempo. La línea naranja representa valores iniciales cercanos a -0.7, mientras que la azul muestra valores alrededor de 0.2 en fechas más recientes. Esto indica un cambio o mejora en la tendencia de los datos entre ambos periodos analizados.
